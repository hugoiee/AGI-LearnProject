# 变量
当执行`a = 'ABC'`时，Python解释器做了两件事：
1. 在内存中创建一个字符串 ABC
2. 在内存中创建一个变量 a ，并指向 ABC 字符串

命名规范：可以包含字母、数字、下划线。但是只能以 字母 和 下划线开头。
PEP8规范：
1. 变量写法： student_name
2. 常量写法： STUDENT_ID
3. 类名写法： HttpRequestHandler

```python
# 字符串
name = "hugo zhao"
age = 18
print(name.title()) # 首字母大写
print(name.upper()) # 全部转大写
print(name.lower()) # 全部转小写
# f-string（格式化字符串字面量），兼容3.6+的版本
print(f"I am {name.title()} and I am {age} years old")
# .format方法，兼容 2.7+的版本，性能略低于 f-string。
print("I am {} and I am {} years old".format(name,age))

print("Languages:\n\tPython\n\tC\n\tJavaScript")

# 去空白
# strip()去两端；lstrip()去左侧；rstrip()去右侧
test_text = " hugo "
print(test_text, len(test_text))
print(test_text.strip(),len(test_text))

test_text = test_text.strip()
print(test_text, len(test_text))

# 删除前缀
starch_url = 'https://nostarch.com'
starch_url = starch_url.removeprefix('https://')
print(starch_url)
```
